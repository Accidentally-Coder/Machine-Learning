#Before we get started ...

Throughout this notebook, we will use  "tips" dataset is a built-in dataset in seaborn that contains information about tips left by customers in a restaurant. It is often used as a sample dataset for demonstrating various data visualization techniques in seaborn and other data analysis libraries.

**Brief overview of the columns in the "tips" dataset:**

**total_bill**: The total bill amount for a meal, including tax and excluding tips.

**tip**: The amount of tip left by the customer.

**sex**: The gender of the customer (male or female).

**smoker**: Whether the customer is a smoker (yes or no).

**day**: The day of the week (e.g., Thursday, Friday, etc.).

**time**: The time of day (Lunch or Dinner).

**size**: The size of the dining party.

##**Loading the dataset**

````python
tips = sns.load_dataset('tips')
````

````python
tips.head()
````

#Part 01 : Distribution Plots

##Displot

Dist plot allows us to show distribution of a univariate set of observations.

```
displot()
```

It visualized through a histogram i.e. only one observation and hence we choose one particular column of the dataset.


````python
sns.displot(tips['total_bill'])
````

````python
sns.displot(tips['total_bill'],kde=True)
````

##Joinplot

Joinplot combines two different plots: a bivariate distribution and another univariate graph.



```
jointplot(x,y,data)
```



````python
sns.jointplot(x='total_bill',y='tip',data=tips)
````

````python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
````

You can also generate regression line(if you are not accustomed with this term, don't worry! We will get acquainted with regression in our upcoming ML basic workshop 😃)

````python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
````

##Pairplot
It plots multiple pairwise bivariate distributions in a dataset,i.e.,it will do jointplot for every single possible combination of the numerical columns in the data frame.



```
pairplot()
```



````python
sns.pairplot(tips)
````

You can also use **hue** parameter which allows to add an additional categorical variable to your pairplot visualization

````python
sns.pairplot(tips,hue='sex')
````

You can also style it with **palette** parameter.

#Part 02 : Categorical Plots

Categorical plots are mainly concerned about visualizing distribution of categorical data.

##Barplot

Bar plots are commonly used to visualize the distribution of a categorical variable and to compare the values of this variable across different categories.

```
barplot(x,y,data)
```



````python
sns.barplot(x='sex',y='total_bill',data=tips)
````

In the above barplot, we have plotted categorical data along x-axis and numerical data along y-axis.

**Note that, by-default *mean* of total_bill for the case of male and female sex have been shown.**

You can change the **estimator** object to showcase your customized distribution.

For instance, let us try to take standard-deviation of total_bills instead of mean.

````python
import numpy as np
````

````python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
````

##Countplot

 It is used to show the counts of observations in each categorical bin using bars.

```
countplot()
```



````python
sns.countplot(x='sex',data=tips)
````

##Boxplot

#Part 03: Matrix Plots

It allows to plot data as color-encoded matrices and can also be used to indicate clusters(will be introduced in ML workshop in details) within the data.

##Heatmap

 It is graphical representation of data using colors to visualize the value of the matrix.

Let's find the correlation of tips dataset.

**If you have just heard the term correlation, this video might be of help...** [correlation](https://youtu.be/G5FkaxWBtkM?feature=shared)

````python
sns.heatmap(tc,annot=True)
````

###Another example of heatmap

````python
pvf = flights.pivot_table(index='month',columns='year',values='passengers')
pvf
````

````python
sns.heatmap(pvf,linecolor='white',linewidths=1)
````

##Clustermap

It uses hierarchal clustering to produce a clustered version of the heatmap.

````python
sns.clustermap(pvf)
````

Basically, it clustered similar years and months together.

Note that the years and months are no longer in order, rather they are grouped by similarity in value (passenger count). That means we can begin to infer things from this plot, such as August and July being similar (makes sense, since they are both summer travel months)
