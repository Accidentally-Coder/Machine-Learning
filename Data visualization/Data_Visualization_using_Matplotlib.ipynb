````python
path = "/content/drive/MyDrive/Dataset/blood_sugar_data.csv"
df = pd.read_csv(path)
df
````

Let's do this project by making a histogram where we will plot sugar-level of male and female separately.

How do we plot histogram? Using hist() method

````python
plt.hist(df['Blood Sugar'],bins=12)
plt.show()
````

**bin controls number of intervals used to create the histogram**

Range = 149-77 = 72,
bin = 12,
interval = 72/12 = 6

Now, back to plotting blood sugar level of male and female separately

````python
male_names = []
male_sugar_level = []

female_names = []
female_sugar_level = []

for index,row in df.iterrows():
  if row['Gender'] == 'Male':
    male_names.append(row['Name'])
    male_sugar_level.append(row['Blood Sugar'])
  elif row['Gender'] == 'Female':
    female_names.append(row['Name'])
    female_sugar_level.append(row['Blood Sugar'])

print(male_names)
print(male_sugar_level)
print(female_names)
print(female_sugar_level)
````

````python
plt.hist([male_sugar_level,female_sugar_level],bins=10,color=["blue","red"])
````

````python
plt.xlabel('sugar range')
plt.ylabel('Frequency of patients')
plt.title('Blood sugar analysis')

plt.hist([male_sugar_level, female_sugar_level], bins=10, color=["blue", "red"],label=["male","female"])
````
